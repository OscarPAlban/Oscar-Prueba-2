## Grafique

```python
# Datos del sistema de ecuaciones
A = np.array([[1, 1], [-2, 5]], dtype=float)
b = np.array([7, 0], dtype=float).reshape(-1, 1)
x0 = np.array([0, 0], dtype=float).reshape(-1, 1)
tol = 1e-5
max_iter = 1000

# Solución y trayectoria usando Gauss-Jacobi
x_jacobi, traj_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)


# Graficar las trayectorias
plt.figure(figsize=(10, 6))

# Trayectoria Gauss-Jacobi
traj_jacobi = np.hstack(traj_jacobi)
plt.plot(traj_jacobi[0, :], traj_jacobi[1, :], 'o-', label='Gauss-Jacobi', color='blue')



# Solución real
plt.plot(5, 2, 's', label='Solución Real', color='green')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.title('Trayectorias de Gauss-Jacobi')
plt.grid(True)
plt.show()
```

# Resuelva

$ x_1 + x_2 = 6$

$  -2 x_1 +  x_2 = 0$

```python
# Datos del sistema de ecuaciones
A = np.array([[1, 1], [-2, 1]], dtype=float)
b = np.array([6, 0], dtype=float).reshape(-1, 1)
x0 = np.array([0, 0], dtype=float).reshape(-1, 1)  # Punto inicial
tol = 1e-5
max_iter = 1000

# Solución y trayectoria usando Gauss-Jacobi
x_jacobi, traj_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)

# Graficar las trayectorias
plt.figure(figsize=(10, 6))

# Trayectoria Gauss-Jacobi
traj_jacobi = np.hstack(traj_jacobi)
plt.plot(traj_jacobi[0, :], traj_jacobi[1, :], 'o-', label='Gauss-Jacobi', color='blue')

# Solución real
sol_real = np.linalg.solve(A, b)  # Usando NumPy para encontrar la solución real
plt.plot(sol_real[0, 0], sol_real[1, 0], 's', label='Solución Real', color='green')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.title('Trayectorias de Gauss-Jacobi y Gauss-Seidel')
plt.grid(True)
plt.show()
```
